```python
x = np.array([ 10 , 8 , 30 , 100 ])
y = np.array([ 2 , 3 ,4  ,5 ])
add=np.add(x,y) #x+y
sub=np.subtract(x,y) #x-y
mul=np.multiply(x,y) #x*y
div=np.divide(x,y) # x/y
remainder = x % y
print(add)
print(sub)
print(mul)
print(div)
print(remainder)

sin_val = np.sin(x)
print(sin_val)

cos_val = np.cos(x)
print(cos_val)

deg_con = np.rad2deg(x)
print(deg_con)
```
