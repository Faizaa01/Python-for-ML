```python
df['Total Marks'] = df.iloc[::, 2:5].sum(axis=1)
df
```

```python
mn = df['Total Marks'].min()
mx = df['Total Marks'].max()
df['Scaled Marks'] = df['Total Marks'].apply(lambda x :(x-mn)/(mx-mn))
df
```

```python
def gradeing(marks):
    if marks>=260:
        return 'A+'
    elif marks>=250:
        return 'A'
    else:
        return 'A-'

df['Grade'] = df['Total Marks'].apply(gradeing)
```

```python
def marking_system(df): 
    a = df['Data Structure Marks'] * 2 
    b = df['Python Marks'] * 3 
    c = df['Algorithm Marks'] * 4

    return a+b+c

df['Ex marks'] = df.apply(marking_system,axis=1)
df
```
