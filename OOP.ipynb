```python
class Phone:
  category="Devices"

  def __init__(self, model, battery, camera, percentage=100):
    self.model=model
    self.battery=battery
    self.camera=camera
    self.percentage=percentage

  def charge(self, hour):
    print(f"charged in {hour}hr")

  def capture(self, photo):
    if(self.percentage < photo):
      print("Phone is dead")
    else:
      self.percentage -= photo
      print(f"photo clicked in {self.model}")

class Cooling_Mechanism:
  def __init__(self, cooling_method):
    self.cooling_method = cooling_method

  def cooling_on(self):
    print(f"System is Cooling by {self.cooling_method}")

apple = Phone("Iphone7",2000,30)
samsung = Phone("s-7", 3000,40)
motorola = Phone("M-12", 3400, 43)

apple.capture(10)
print(apple.percentage)
```

```python
class Smart(Phone):
  def __init__(self, model, battery, camera, processor):
    super().__init__(model, battery, camera)
    self.processor=processor

  def charge(self, hour):
    print("fast Charging")
    super().charge(hour)
pro = Smart("Sxs",5000,40,"snapDragon")
pro.charge(3)
```

```python
class Smart_Cool(Smart, Cooling_Mechanism):
  def __init__(self, model, battery, camera, processor, cooling_method):
    Smart.__init__(self, model, battery, camera, processor)
    Cooling_Mechanism.__init__(self, cooling_method)

pro_cooling = Smart_Cool("Sxs", 5000, 40,"snapDragon", "Water")
print(pro_cooling.model)
print(pro_cooling.processor)
print(pro_cooling.cooling_method)
print(pro_cooling.cooling_on())
```

```python
class Camera:
  def __init__(self, name):
    self.name = name

  def capture(self):
    print("Photo is done for")

class Phone(Camera):
  def __init__(self, name, resolution):
    super().__init__(name)
    self.resolution = resolution
  def capture(self):
    print("Photo is done for phone")

class Drone(Camera):
  def __init__(self, name, resolution):
    super().__init__(name)
    self.resolution = resolution
  def capture(self):
    print("Photo is done for drone")

class DSLR(Camera):
  def __init__(self, name, resolution):
    super().__init__(name)
    self.resolution = resolution
  def capture(self):
    print("Photo is done for DSLR")

Phone("Redmi",400).capture()
Drone("DJI",400).capture()
DSLR("Sony",400).capture()
```

```python
from inspect import modulesbyfile
class Mobile():
  def __init__(self, model, imei):
    self.__model = model
    self.__imei = imei

  def imei_getter(self):
    return self.__imei
  def model_getter(self):
    return self.__model
  def model_setter(self, model):
    self.__model = model

phone = Mobile("iPhone","12m2e2n")
print(phone.imei_getter())
print(phone.model_getter())
phone.model_setter("Samsung")
print(phone.model_getter())
```

```python
from abc import ABC, abstractmethod
class Telephone(ABC):
  @abstractmethod
  def make_call(self):
    print("Making a call")

  def capture(self):
    pass

class Sphone(Telephone):
  def make_call(self):
    print("Making a call")

class Iphone(Telephone):
  def make_call(self):
    print("Making a call")

ip = Iphone()
ip.make_call()
```
